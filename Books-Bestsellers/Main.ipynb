# Bestsellers EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing Datasets

```python
df = pd.read_csv('bestsellers with categories.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

# 1. User Ratings

```python
df['User Rating'].max()
```

## > Top Rated

```python
top_ratings = df.sort_values('User Rating',ascending=False)
```

```python
top_ratings.head(10)
```

## > Least Rated

```python
bottom_ratings = df.sort_values('User Rating')
```

```python
bottom_ratings.head()
```

## Observations :
